```python
from sympy import *
```

```python
x,y,z = symbols('x y z', real=True)
```

```python
f = -2*x**4 + 4*x**2*exp(y) - 4*exp(4*y)
f
```

```python
fx = diff(f,x)
fy = diff(f,y)
fxx = diff(fx,x)
fyy = diff(fy,y)
fxy = diff(fy,x)
fx,fy,fxy,fxx,fyy
```

```python
eq1 = Eq(fx,0)
eq2 = Eq(fy,0)
eq1,eq2
```

```python
sols = solve([eq1,eq2],[x,y])
sols
```

```python
D = fxx*fyy - fxy**2
D
```

```python
for s in sols:
    print(s,D.subs([(x,s[0]),(y,s[1])]),fxx.subs([(x,s[0]),(y,s[1])]))
```

```python
from sympy.plotting import plot3d
```

```python
rad =0.1
for s in sols:
    a = s[0]
    b = s[1]
    plot3d(f,(x,a-rad,a+rad),(y,b-rad,b+rad))
```
